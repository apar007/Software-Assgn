```python
df['old_locus_tag(gene)'] = df['old_locus_tag(gene)'].apply(lambda x: np.nan if str(x).startswith('locus_tag') else x)
```

```python

condition = df['codonStart'].str.startswith('product', na=False)


df.loc[condition, 'product'] = df.loc[condition, 'codonStart']


df.loc[condition, 'codonStart'] = 'codon_start=1'


print(df[['codonStart', 'product']])
```

```python
condition = df['translTable'].str.startswith('protein_id', na=False)


df.loc[condition, 'protein _id'] = df.loc[condition, 'translTable']


df.loc[condition, 'translTable'] = 'transl_table=11'
```
